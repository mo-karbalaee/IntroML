## 1.1. Gaussian Noise

```python
noisy_gaussian = add_gaussian_noise(image=image, mean=10, sigma=80)
display_images(image, noisy_gaussian, title="Gaussian Noise")
```

## 1.2. Poisson Noise

```python
noisy_poisson = add_poisson_noise(image=image)
display_images(image, noisy_poisson, title="Poisson Noise")
```

## 1.3. Salt and Pepper Noise

```python
noisy_salt_pepper = add_salt_and_pepper_noise(image=image, salt_prob=0.01, pepper_prob=0.1)
display_images(image, noisy_salt_pepper, title="Salt and Pepper Noise")
```
